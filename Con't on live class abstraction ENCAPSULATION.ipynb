```python
class earth:
    def __init__ (self, a, b, c):
        self.a = a 
        self.b = b 
        self.c = c
        
    def __str__(self):
        return "this is the the return from my earth class for ENCAPSULATION"
    
class moon:
    def __init__ (self, d, e, f):
        self.d = d
        self.e = e 
        self.f = f
        
    def __str__(self):
        return "this is the the return from my moon class for ENCAPSULATION"
    
class star:
    def __init__ (self, g, h, i):
        self.g = g
        self.h = h 
        self.i = i
        
    def __str__(self):
        return "this is the the return from my star class for ENCAPSULATION"
    
class sun:
    
    def __init__(self, earth, moon, star):
        self.earth = earth
        self.moon = moon
        self.star = star
    
    def __str__(self):
        return "this is what encompasses all of the above mentioned for my ENCAPSULATION" + "  "  + str(self.earth) + "  " + str(self.moon) + "  " + str(self.star)
```
